Solving 1d diffusion equation
$$ \partial_t \phi = \kappa \frac{\partial^2 \phi}{\partial x^2} $$
Stability condition:
$$ \Delta t < \frac{h^2}{4\kappa} $$
time forward done using **RK2**
and using 3 point central difference


1. For 0 boundary conditions:

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return exp(-4*(x-2)**2) 
N=512  #no. of data points
l=10.0 
h=l/N
tf=4.0  
dt=0.0001
x=linspace(0.0, 10.0, N)
f=zeros(N+2)

f[1:N+1]=fun(x[0:N])
f[0]=0; f[N+1]=0
f_mid=zeros(N+2)
t=0.0
mark=0.0
times=[0.0, 1.0, 2.0, 3.0, 4.0]
pos=0
c = 0.5
while (t<=tf+dt/2):
    f_mid[1:N+1]=f[1:N+1] + (c*dt/(h*h))*(f[2:N+2]-2*f[1:N+1]+f[0:N])/2.0
    f_mid[0] = 0; f_mid[-1] = 0
    
    f[1:N+1] = f[1:N+1] + (c*dt/(h*h))*(f_mid[2:N+2]-2*f[1:N+1]+f_mid[0:N])
    f[0] = 0; f[-1] = 0
    if(t+dt/2>times[pos]):
        plt.plot(x,f[1:N+1],label=times[pos])
        pos+=1
    t+=dt
plt.legend(loc=0)
plt.xlabel(r'$x$')
plt.ylabel(r'$\Psi(x, t)$')
plt.title('1D Diffusion Equation')
plt.show()
```

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return exp(-4*(x-2)**2) 
N=512  #no. of data points
l=10.0 
h=l/N
tf=4.0  
dt=0.0001
x=linspace(0.0, 10.0, N)
f=zeros(N+2)

f[1:N+1]=fun(x[0:N])
f[0]=f[N]; f[N+1]=f[1]
f_mid=zeros(N+2)
t=0.0
mark=0.0
times=[0.0, 1.0, 2.0, 3.0, 4.0]
pos=0
c = 0.5
while (t<=tf+dt/2):
    f_mid[1:N+1]=f[1:N+1] + (c*dt/(h*h))*(f[2:N+2]-2*f[1:N+1]+f[0:N])/2.0
    f_mid[0] = f_mid[N]; f_mid[-1] = f_mid[1]
    
    f[1:N+1] = f[1:N+1] + (c*dt/(h*h))*(f_mid[2:N+2]-2*f[1:N+1]+f_mid[0:N])
    f[0] = f[N]; f[-1] = f[1]
    if(t+dt/2>times[pos]):
        plt.plot(x,f[1:N+1],label=times[pos])
        pos+=1
    t+=dt
plt.legend(loc=0)
plt.xlabel(r'$x$')
plt.ylabel(r'$\Psi(x, t)$')
plt.title('1D Diffusion Equation')
plt.show()
```
