Solving Wave Eqution
$$ \partial_t \phi + c \partial_x \phi =0 $$
This is the wave eqution for traveling **rightwards**.  Also called **advection eqution**. 

 Using **Central Difference** method gives unstable solution. 
 
 We use **Upwind Scheme**, for **c>0**, i.e. a wave traveling rightwards. Uses Backward difference for x derivative. i.e.  $\frac{ \phi_i-\phi_{i-1}}{h}$
But it is dissipative

 And for **c < 0**, We use forward deriavative. i.e. $\frac{ \phi_{i+1}-\phi_{i}}{h}$ 
 
 Following code uses RK2 method and central derivative

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return exp(-16*(x-2)**2)

N=625
l=10.0
h=l/(N)
tf=4.0
dt=0.001
x=linspace(0.0, 10.0, N)
f=zeros(N+2)
f[1:N+1]=fun(x)
f[0]=0; f[N+1]=0
f_mid=zeros(N+2)
t=0.0
mark=0.0
c = 1.5
while (t<=tf+dt/2):
    f_mid[1:N+1]=f[1:N+1] - (c*dt/(2*h))*(f[2:N+2]-f[0:N])/2.0
    # f_mid[0] = f_mid[N]; f_mid[-1] = f_mid[1]
    
    f[1:N+1] = f[1:N+1] - (c*dt/(2*h))*(f_mid[2:N+2]-f_mid[0:N])
    # f[0] = f[N]; f[-1] = f[1]
    if(t+dt/2>mark):
        plt.plot(x,f[1:N+1],label=mark)
        mark+=1.0
    t+=dt
plt.legend(loc=0)
plt.show()
```

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return exp(-16*(x-2)**2)

N=100
l=10.0
h=l/(N-1)
tf=4.0
dt=0.001
x=linspace(0.0, 10.0, N)
f=zeros(N+2)
f[1:N+1]=fun(x)
f[0]=0; f[N+1]=0
f_mid=zeros(N+2)
t=0.0
mark=0.0
c = 1.5
while (t<=tf+dt/2):
    f_mid[1:N+1]=f[1:N+1] - (c*dt/(h))*(f[1:N+1]-f[0:N])/2.0
    # f_mid[0] = f_mid[N]; f_mid[-1] = f_mid[1]
    
    f[1:N+1] = f[1:N+1] - (c*dt/(h))*(f_mid[1:N+1]-f_mid[0:N])
    # f[0] = f[N]; f[-1] = f[1]
    if(t+dt/2>mark):
        plt.plot(x,f[1:N+1],label=mark)
        mark+=1.0
    t+=dt
plt.legend(loc=0)
plt.show()
```

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return exp(-32*(x-2)**2)

N=500
l=2*pi
h=l/(N)
tf=1.0
dt=0.001
x=linspace(0.0, l, N)
f=zeros(N+2)
f[1:N+1]=fun(x)
f[0]=f[N]; f[N+1]=f[1]
f_mid=zeros(N+2)
t=0.0
mark=0.0
c = 1.5
while (t<=tf):
    f_mid[1:N+1]=f[1:N+1] - (c*dt/(2*h))*(f[1:N+1]-f[0:N])
    f_mid[0] = f_mid[N]; f_mid[-1] = f_mid[1]
    
    f[1:N+1] = f[1:N+1] - (c*dt/(h))*(f_mid[1:N+1]-f_mid[0:N])
    f[0] = f[N]; f[-1] = f[1]
    if(t+dt/2>mark):
        plt.plot(x,f[1:N+1],label=mark)
        mark+=1.0
    t+=dt
plt.legend(loc=0)
plt.show()
for i in range(N):
    if(round(x[i],2)==3.48):
        print(round(f[i],2))
        
```
