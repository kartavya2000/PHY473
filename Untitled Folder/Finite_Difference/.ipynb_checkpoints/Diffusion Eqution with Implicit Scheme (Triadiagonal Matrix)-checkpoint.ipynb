Diffusion Equation using Implicit Scheme

RK2 for time forward

1. For 0 boundary conditions

```python
from numpy import *
import matplotlib.pyplot as plt
import scipy.sparse as sps
from scipy import linalg

def fun(x):
    return exp(-4*(x-2)**2) 
N=100  #no. of data points
l=10.0 
h=l/N
tf=4.0  
dt=0.01
x=linspace(0.0, 10.0, N)
f=zeros(N+2)

f[1:N+1]=fun(x[0:N])
f[0]=0; f[N+1]=0
f_mid=zeros(N+2)
t=0.0
mark=0.0
times=[0.0, 1.0, 2.0, 3.0, 4.0]
pos=0
c = 0.5
while (t<=tf+dt/2):
    X=(1+c*dt/h**2); Y=-c*dt/(2*h**2)
    A=(sps.diags([Y, X, Y], [-1, 0, 1], shape=(N, N), format='csc'))
    A=A.toarray()
    f[1:N+1]=dot(linalg.inv(A), (f[1:N+1]+c*dt/(2.0*h**2) * (f[2:N+2]-2*f[1:N+1]+f[0:N])))
    
    if(t+dt/2>times[pos]):
        plt.plot(x,f[1:N+1],label=times[pos])
        pos+=1
    t+=dt
plt.legend(loc=0)
plt.xlabel(r'$x$')
plt.ylabel(r'$\Psi(x, t)$')
plt.title('1D Diffusion Equation')
plt.show()
```

2. For periodic boundary conditions

```python
from numpy import *
import matplotlib.pyplot as plt
import scipy.sparse as sps
from scipy import linalg

def fun(x):
    return exp(-4*(x-2)**2) 
N=100  #no. of data points
l=10.0 
h=l/N
tf=4.0  
dt=0.01
x=linspace(0.0, 10.0, N)
f=zeros(N+2)

f[1:N+1]=fun(x[0:N])
f[0]=f[N]; f[N+1]=f[1]

t=0.0
mark=0.0
times=[0.0, 1.0, 2.0, 3.0, 4.0]
pos=0
c = 0.5
while (t<=tf+dt/2):
    X=(1+c*dt/h**2); Y=-c*dt/(2*h**2)
    A=(sps.diags([Y, X, Y], [-1, 0, 1], shape=(N, N), format='csc'))
    A=A.toarray()
    A[0, N-1]=Y
    A[N-1, 0]=Y
    f[1:N+1]=dot(linalg.inv(A), (f[1:N+1]+c*dt/(2.0*h**2) * (f[2:N+2]-2*f[1:N+1]+f[0:N])))
    f[0]=f[N]; f[N+1]=f[1]

    if(t+dt/2>times[pos]):
        plt.plot(x,f[1:N+1],label=times[pos])
        pos+=1
    t+=dt
plt.legend(loc=0)
plt.xlabel(r'$x$')
plt.ylabel(r'$\Psi(x, t)$')
plt.title('1D Diffusion Equation')
plt.show()
```
