Burger's Equation:
$$ \partial_t u + u\partial_x u = \nu \partial_{x}^2 u $$

Euler forward employed for moving forward in time... Alternatively RK2 can be used.
For nonlinear term($u \partial_t x u$) upwind scheme of derivative is used and for diffusion term ($\nu \partial_{x}^2 u$) 
Diffusion term has time scale $h^2/\nu$ and $h/u_{rms}$ for the nonlinear term. h is the grid size. Choose smaller of the two time scales as $\Delta t$

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return sin(x)

N=128
l=pi*2
h=l/(N)
tf=1.0
dt=0.001
x=linspace(0.0, l, N)
f=zeros(N+2)
f[1:N+1]=fun(x)
f[0]=0; f[N+1]=0
# f_mid=zeros(N+2)
t=0.0
mark=0.0
v = 0.1
th=1
while (t<=tf+dt/2):
    nf=zeros(N)
    
    for i in range(1, N+1):
        if(f[i]>0):
            nf[i-1] = f[i] - dt/h * f[i]*(f[i]-f[i-1]) + v*dt/h**2 *(f[i+1] - 2.0*f[i] + f[i-1])
        else:
            nf[i-1] = f[i] - dt/h * f[i]*(f[i+1]-f[i]) + v*dt/h**2 *(f[i+1] - 2.0*f[i] + f[i-1])
    f[1:N+1] = nf[0:N+1]
    if(t+dt/2>mark):
        plt.plot(x,f[1:N+1],label=mark, lw=th)
        th+=.1
        mark+=0.2
    t+=dt
plt.legend(loc=0)
plt.show()
```
