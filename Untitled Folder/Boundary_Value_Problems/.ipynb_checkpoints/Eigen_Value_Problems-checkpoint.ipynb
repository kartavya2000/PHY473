Soving Eigen Value Problems of the form:
$$ y'' = \lambda y $$
In order to solve Secant Method is employed for convergence and Odeint is used for moving forward. Alternatively Euler Forward can also be used and the same is demonstrated in the second part. 

As is usually the case in secant method, we must take the initial guess values to be on the same side such that a monotonous sequence  converges to the required value.

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def ydot(y, x):
    return (y[1], a*y[0])
N = 100
x = linspace(0.0, 2.0, N)
slope_a = 0.1 #initial slope can be taken any value, 
                #afterwards impose normalization condition
yinit = array([0, slope_a]) 
a = -2
y = odeint(ydot, yinit, x)
yfinal = y[-1, 0]
a_p = a
a = -1
tol = 0.001
iter = 0
while((abs(a-a_p)>tol) and (iter<10)):
    y = odeint(ydot, yinit, x)
    yfinal_p, yfinal = yfinal, y[-1, 0]
    a_p, a = a, a - yfinal*(a-a_p)/(yfinal-yfinal_p)
    iter +=1
plt.plot(x,y[:,0])
```

```python
from numpy import *
import matplotlib.pyplot as plt

def ydot(y, x):
    return matrix([y[1], a*y[0]])
def EulerForward(ydot, yinit, x):
    n = len(x)
    y = zeros((n, len(yinit)))
    dx = x[1]-x[0]
    y[0,:] = yinit[:]
        
    for i in range(1, n):
            y[i, :] = y[i-1, :] + ydot(y[i-1, :], x[i-1]) * dx
    return y

N = 100
x = linspace(0.0, 2.0, N)
slope_a = 0.1 #initial slope can be taken any value, 
                #afterwards impose normalization condition
yinit = array([0, slope_a]) 
a = -2
y = EulerForward(ydot, yinit, x)
yfinal = y[-1, 0]
a_p = a
a = -1
tol = 0.001
iter = 0
while((abs(a-a_p)>tol) and (iter<10)):
    y = EulerForward(ydot, yinit, x)
    yfinal_p, yfinal = yfinal, y[-1, 0]
    a_p, a = a, a - yfinal*(a-a_p)/(yfinal-yfinal_p)
    iter +=1
plt.plot(x,y[:,0])
```
