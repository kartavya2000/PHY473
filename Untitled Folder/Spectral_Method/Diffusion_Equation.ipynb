Diffusion Equation using Spectral Method. 

Time stepping done with Euler Forward

```python
import matplotlib.pyplot as plt
from numpy import *
dt = 0.0001
tf = 0.1100
l=20.0
def f(x):
    return sin(pi*x)+ sin(2*pi*x)
n=256
dx=l/n
x=arange(0,n)*dx
y=f(x)
yt=fft.fft(y,n)/n
t=0.0
mark = 0.0
kx=fft.fftfreq(n, l/(2*pi*n))
while (t<=tf):
    yt*=(1.0-kx*kx*dt)
    if(mark<t+dt/2.0):
        y=fft.ifft(yt, n)*n
        plt.plot(x,real(y), label=round(mark,2))
        mark+=0.02
    t+=dt
plt.legend(loc=0)
plt.title(r'$Diffusion\ Equation$')
plt.xlabel(r'$x$')
plt.ylabel(r'$Re(\rho(x,t))$')
plt.show()
```

Time Stepping done with RK2 method:

```python
import matplotlib.pyplot as plt
from numpy import *
dt = 0.0001
tf = 0.1100
l=20.0
def f(x):
    return sin(pi*x)+ sin(2*pi*x)
n=256
dx=l/n
x=arange(0,n)*dx
y=f(x)
yt=fft.fft(y,n)/n
t=0.0
mark = 0.0
kx=fft.fftfreq(n, l/(2*pi*n))
while (t<=tf):
    ytm=yt-kx*kx*dt*0.5*yt
    yt=yt-kx*kx*dt*ytm
    if(mark<t+dt/2.0):
        y=fft.ifft(yt, n)*n
        plt.plot(x,real(y), label=round(mark,2))
        mark+=0.02
    t+=dt
plt.legend(loc=0)
plt.title(r'$Diffusion\ Equation$')
plt.xlabel(r'$x$')
plt.ylabel(r'$Re(\rho(x,t))$')
plt.show()
```
