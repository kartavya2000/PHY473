Wave Equation by Spectral Method

Using Euler Forward Time stepping

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return (cos(x-2.0)**2)*exp(-32.0*(x-2)**2)

n=256
l=5.0
c=0.5
dt=0.0001
tf=4.0
h=l/n
x=arange(0, n)*h
k=fft.fftfreq(n, l/(2.0*n*pi))
y=fun(x)
i=complex(0.0, 1.0)
a=2*pi/l
yt=fft.fft(y, n)
t=0.0
mark=0.0
while(t<=tf+dt/2):
    yt=yt*(1-i*c*k*dt)
    if(t+dt/2>mark):
        y=fft.ifft(yt)
        plt.plot(x,real(y),label=mark)
        mark+=1.0
    t+=dt
plt.legend(loc=0)
plt.show()
```

Using RK2 time stepping

```python
from numpy import *
import matplotlib.pyplot as plt

def fun(x):
    return (cos(x-2.0)**2)*exp(-32.0*(x-2)**2)

n=256
l=5.0
c=0.5
dt=0.0001
tf=4.0
h=l/n
x=arange(0, n)*h
k=fft.fftfreq(n, l/(2.0*n*pi))
y=fun(x)
i=complex(0.0, 1.0)
a=2*pi/l
yt=fft.fft(y, n)
t=0.0
mark=0.0
while(t<=tf+dt/2):
    ytm=yt*(1-i*c*k*dt*0.5)
    yt=yt-ytm*i*c*k*dt
    if(t+dt/2>mark):
        y=fft.ifft(yt)
        plt.plot(x,real(y),label=mark)
        mark+=1.0
    t+=dt
plt.legend(loc=0)
plt.show()
```
