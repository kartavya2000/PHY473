```python
from numpy import *
import matplotlib.pyplot as plt
j = complex(0.0, 1.0)
def f(x):
    return 1/sqrt(2)*(pi**-0.25 * exp(-0.5*x*x)) * (1 + sqrt(2)*x)
N = 101
L = 10.0*pi
h = L/(N)
def V(x):
    return 0.25 * x**4
dt = 0.0001
x = linspace(-L/2.0, L/2.0, N)
R = zeros(N+2)
I = zeros(N+2)

R[1:N+1] = f(x[0:N])

R[0] = R[N]; R[N+1] = R[1];
I[0] = I[N]; I[N+1] = I[1];

I[1:N+1] = I[1:N+1] - 0.5 * dt *(-0.5 * (h**-2)  * (R[0:N] - 2.0*R[1:N+1] + R[2:N+2]) + V(x)*R[1:N+1])
I[0] = I[N]; I[N+1] = I[1];

mark = [2.0]
tf = 2.0
# mark = linspace(0.0, tf, 5)
t = 0.0
pos= 0

while(t<=tf+dt/2.0):
    I1 = zeros(N+2)
    R[1:N+1] += dt * (-0.5 * (h**-2) * (I[0:N] - 2.0*I[1:N+1] + I[2:N+2]) + V(x)*I[1:N+1])
    R[0] = R[N]; R[N+1] = R[1];
    I1[1:N+1] = I[1:N+1] - 0.5 * dt * (-0.5 * (h**-2) * (R[0:N] - 2.0*R[1:N+1] + R[2:N+2]) + V(x)*R[1:N+1])
    I[1:N+1] += -dt * (-0.5 * (h**-2) *(R[0:N] - 2.0*R[1:N+1] + R[2:N+2]) + V(x)*R[1:N+1])
    I[0] = I[N]; I[N+1] = I[1];
    if (t+0.5*dt>=mark[pos]):
        plt.plot(x, R[1:N+1]**2 + I1[1:N+1]**2, label = "$t={mark}$".format(mark=mark[pos]))
        break
    t += dt
plt.legend(loc = 0, fontsize = 20)
plt.xlabel(r'$x$', size = 20)
plt.ylabel("$\psi (x, t)$", fontsize = 20)
plt.title("1D Schoridnger Equation", fontsize=20)
plt.grid()
plt.show()
for xc in R[1:N+1]**2 + I[1:N+1]**2:
    print(round(xc, 5))
```
