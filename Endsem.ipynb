```python
from numpy import *

def ydot(y, x):
    return matrix([y[1], -y[0]])
def EulerForward(ydot, yinit, x):
    n = len(x)
    y = zeros((n, len(yinit)))
    dx = x[1]-x[0]
    y[0,:] = yinit[:]
        
    for i in range(1, n):
            y[i, :] = y[i-1, :] + ydot(y[i-1, :], x[i-1]) * dx
    return y
dt=0.1
x=arange(0,2*pi+dt, 0.1)
yinit=array([0,1])
y=EulerForward(ydot, yinit, x)
for i in range(len(x)):
    if(round(x[i],2)==round(2*pi/7,2)):
        print(y[0,i])
```
